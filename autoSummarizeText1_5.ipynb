```python
import re
import urllib2
from bs4 import BeautifulSoup
```

```python
articleURL = urllib2.urlopen('https://timesofindia.indiatimes.com/new-to-canada/immigration-citizenship/Canada-a-home-away-from-home/articleshow/48710982.cms')
soup = BeautifulSoup(articleURL)
data = soup.findAll(text=True)
data
```

```python
mydivs = soup.findAll("div", {"class": "Normal"})
mydivs
```

```python
from bs4.element import Comment
```

```python
 def visible(element):
    if element.parent.name in ['style', 'script', '[document]', 'head', 'title', 'meta']:
        return False
    elif re.match('<!--.*-->', str(element.encode('utf-8'))):
        return False
    return True
```

```python
def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True
```
