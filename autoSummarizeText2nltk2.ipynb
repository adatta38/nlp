```python
import urllib2
from bs4 import BeautifulSoup
```

```python
articleURL = "https://timesofindia.indiatimes.com/new-to-canada/immigration-citizenship/Canada-a-home-away-from-home/articleshow/48710982.cms"
```

```python
page = urllib2.urlopen(articleURL).read()
soup = BeautifulSoup(page)
soup
```

```python
# kill all script and style elements
for script in soup(["script", "style", "head", "title", "meta", "[document]"]):
    script.extract()    # rip it out

print(script.extract())
```

```python
# get text
text = soup.get_text()

print(text)
```

```python
mydivs = soup.findAll("div", {"class": "Normal"})
mydivs
```

```python
# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
# lines = (line.strip() for line in mydivs.splitlines())

print(lines)
```

```python
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))

print(chunks)
```

```python
# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

print(text.encode('utf-8'))
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sentences = sent_tokenize(text)
# sentences = sent_tokenize(mydivs)
sentences
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords
```

```python
word_sent = [word for word in word_sent if word not in _stopwords]
word_sent
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict
ranking = defaultdict(int)

for i,sent in enumerate(sentences):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]

ranking
```

```python
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx
```

```python
[sentences[j] for j in sorted(sents_idx)]
```
