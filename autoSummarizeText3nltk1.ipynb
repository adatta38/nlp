```python
import urllib2
from bs4 import BeautifulSoup
```

```python
import lxml
```

```python
articleURL = "https://timesofindia.indiatimes.com/business/international-business/coca-cola-buys-coffee-chain-costa-for-5-1-billion/articleshow/65622023.cms"
```

```python
page = urllib2.urlopen(articleURL).read().decode('utf8', 'ignore')
soup = BeautifulSoup(page, 'html.parser')
soup
```

```python
mydivs = soup.find("div", {"class": "section1"})
mydivs
```

```python
onlyText = mydivs.get_text(strip=True)
onlyText
```

```python
text = onlyText.replace("\t", "").replace("\r", "").replace("\n", "").replace("``", "").replace(u'\xa0', u' ')
text
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sentences = sent_tokenize(text)
sentences
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords
```

```python
word_sent = [word for word in word_sent if word not in _stopwords]
word_sent
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict
```

```python
ranking = defaultdict(int)

for i,sent in enumerate(sentences):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]

ranking
```

```python
sents_idx = nlargest(3, ranking, key=ranking.get)
sents_idx
```

```python
[sentences[j] for j in sorted(sents_idx)]
```
